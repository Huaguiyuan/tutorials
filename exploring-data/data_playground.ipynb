```python
import statsmodels.api as sm

X = df['input0']
y = df['output']

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

```python
X = df['input1']
y = df['output']

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

```python
X = df[['input0','input1']]
y = df['output']

model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```

```python
X = df[['input0', 'input1']]
y = df['output']

X = sm.add_constant(X)
model = sm.OLS(y, X).fit()
predictions = model.predict(X)

model.summary()
```
